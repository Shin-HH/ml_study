### An introduction to additive modeling
> **additive modeling**: 보다 복잡한 표현(함수, 모델)을 만들기 위해 single term들을 추가해 나가는 것  

복잡한 함수를 만났을 때, 그 함수 전체를 생각해보는 것보다 각각의 항들을 떠올리는 것이 더 쉽기 때문에 additive modeling은 유용한 기술이 된다.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-30, 31)
y = 30 + x + np.sin(x)

plt.figure()
plt.plot(x, y, label = r'$y = ?$')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
plt.legend();
```

> 위와 같은 그래프가 있고, 함수 $f(x)$는 모르는 상황

위의 함수가 몇 개의 single term으로 구성 되어있고 그것들을 맞춰야한다고 가정해보자. 우리는 1개의 single term을 더하고 현재의 함수와 target 함수의 차이를 고려해서 다음에 추가할 term을 다시 생각할 수 있다.  
그래프의 y절편이 30이니까 **$y = 30$**이라는 함수로 target함수를 추정해보자.

```python
x = np.arange(-30, 31)
y_hat = np.zeros_like(x) + 30

plt.figure()
plt.plot(x, y_hat, label = r'$y = 30$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend();
```

다음으로 그래프의 기울기를 맞추기 위해서 45도선 즉, **$y=x$** 함수를 추가해보자.

```python
x = np.arange(-30, 31)
y_hat = x + 30

plt.figure()
plt.plot(x, y_hat, label = r'y = 30 + x')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend();
```

원래 함수와 꽤나 비슷해진 것을 볼 수 있다. 하지만 target 함수의 구불구불함은 아직 표현해내지 못하고 있으니, 우리의 함수에 sin을 더해주자.

```python
x = np.arange(-30, 31)
y_hat = x + 30 + np.sin(x)

plt.figure()
plt.plot(x, y_hat, label = r'$y = 30 + x + \sin{x}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.legend();
```

복잡한 함수를 간단한 부분적인 함수들로 분해하는 것은 divide and conquer strategy와 같다. 우리가 원하는 target 함수를 **$F(x)$**라고 한다면 **$y = F(x) = 30 + x + \sin{x}$**이고, 이 함수를 1개의 항으로 이루어진 함수들로 나눈다면 아래와 같이 쓸 수 있다.
> $F(x) = f_1(x) + f_2(x) + f_3(x)   
\\ where,\\ f_1(x) = 30,\\ f_2(x) = x,\\ f_3(x) = \sin{x}$
  
더 일반적으로 표현하면 **하나의 함수의 분해는 M개의 부분적인 함수들의 합으로 나타낼 수 있다.**
> $F_m(x) = f_1(x) + \cdots + f_M(x) = \sum_{m=1}^{M}{f_m(x)}$
  
머신러닝에서는 연속적인 함수보다는 $(x, y)$의 데이터를 받는다. 우리의 목표는 데이터 포인트를 잘 설명하는 함수를 만들어 내는 것이고 우리는 이 함수를 model이라고 부른다. 하나의 완전한 함수를 만들기 위해 부분적인 함수들을 더해나가는 방법을 additive modeling이라고 한다. **Gradient Boosting Machine은 모델을 만들기 위해 아주 심플한 submodel들을 더해가는 additive modeling을 활용하여 점차적으로 모델을 좋은 모델이 되게끔 유도한다.**

### An introduction to boosted regression
**부스팅은 하나의 composite model을 만들기 위해 여러개의 단순한 model들을 결합하는 전략이다.** 간단한 model을 추가할 수록 전체적인 모델은 성능이 더 좋아진다는 아이디어에서 시작되었고, 이 간단한 model을 weak models 또는 weak learner라고 부른다.  
회귀 모델을 학습시킨다는 것은 가능한 한 데이터를 잘 설명하는 함수를 추정하는 것과 같다. composite model의 예측($\hat{y}$)은 M개의 weak model들의 합($\hat{y} = \sum_{m=1}^{M}{f_m(x)}$)으로 표현할 수 있다. 여기서 weak model은 K-NN, regression tree등 어떤 model도 될 수 있다. 하지만 대부분의 경우에 의사결정나무를 부스팅에 사용한다.
  
종종 additive model의 각각의 항 $f_m(x)$들을 독립적이고 병렬적으로 만들 수도 있지만, 부스팅의 경우에는 불가능하다. 부스팅은 전체 모델의 성능을 상승시키기 위해 weak model을 한 단계에 하나씩 학습시켜 더한다. 부스팅은 greedy한 전략을 가지고 있어, 각 단계에서 어떤 모델을 선택하더라도 이미 더해진 모델을 바꾸지 않는다.
  
$\hat{y} = F_M(x)$의 성능이 충분히 좋거나 $f_m(x)$가 더이상 아무 것도 더하지 않을 때 weak model을 더하는 것을 멈출 수 있으나, 일반적으로 M이라는 하이퍼파라미터를 정해서 단계의 수를 결정할 수 있다. M이 높으면 과대적합될 위험이 커진다는 것을 기억하자.

> $F_m(x) = F_{m-1}(x) + f_m(x)$
  
위의 식은 다음 모델을 얻기 위해 현재의 모델을 수정해야하는 것을 보여준다.
  
부스팅은 weak leaner의 형태를 정하지 않는다. 하지만 weak model은 meta model의 형태를 정하게 된다. 예를 들어 weak model들이 선형모형들이었다면, meta model 또한 선형 모형이된다. 또 만약 weak model을 아주 작은 regression tree를 사용했다면 meta model은 나무들의 숲처럼 될 것이다.

### The intuition behind gradient boosting
부스팅 회귀 모델을 만들기 위해 $f_0(x)$라는 형편없는 모델을 만들고, overall model $F_m(x)$가 실제 값인 $y$에 점차적으로 접근할 수 있도록 수정해보자.
> $\begin{align} \hat{y} &= f_0(x) + \Delta_1(x) + \Delta_2(x) + \cdots + \Delta_M(x) \\ &= f_0(x) + \sum_{m=1}^{M}{}\Delta_m(x)\\ &=F_M(x) \end{align}$  
>  
> 또는  
>  
> $F_0(x) = f_0(x) \\ F_m(x) = F_{m-1}(x) + \Delta_m(x) $

  
골프 선수는 첫번째 샷을 치고난 후 $y$와 첫번째 근사의 차이를 통해 즉, $y - f_0(x)$를 통해 다음으로 칠 공의 방향과 힘을 결정한다. 이런 실제 값과 예측 값의 차이를 residual(잔차)이라고 한다.   

위와 같은 상황은 $\Delta_1(x)$를 잔차$y - f_0(x)$에 대해 학습시키는 것과 같다.
  


  
GBM은 학습률 $\eta$가 존재해서 $\hat{y}$가 $y$에 근사라는 것의 속도를 올리거나 내릴 수 있다. 또한 이 학습률은 과적합의 가능성을 줄여준다.

##### Intuitions
- weak model들은 단순한 양 벡터가 아닌 방향 벡터를 학습한다.
- 초기 모델 $f_0(x)$는 $x$를 가지고 $y$를 학습하고자 하지만, $\Delta_m(x)$(초기 모델이 아닌 weak model들)은 $x$를 가지고 방향 벡터를 학습하고자 한다.
- 모든 weak model들은 $x$의 함수이다.
- 두가지 일반적인 방향 벡터는 residual($y - F_{m-1}(x)$)과 sign($sign(y - F_{m-1}(x))$)이다.

### Gradient boosting regression by example

```python
import pandas as pd
from IPython.display import Markdown, display
sqfeet = np.array([750, 800, 850, 900, 950]) # feature
rent = np.array([1160, 1200, 1280, 1450, 2000]) # target

dataset = pd.DataFrame({
    'sqfeet' : sqfeet,
    'rent' : rent
})
display(Markdown('### Orignal Dataset'), dataset)
```

$f_0(x)$는 y의 평균으로 예측하는 모델
- MSE를 최소화시키는 상수는 평균이기 때문이다.

```python
f_0 = np.mean(rent)
F_0 = f_0
residual = rent - F_0

dataset['$F_0$'] = F_0
dataset['y - $F_0$'] = residual

display(Markdown('### $F_0(x)$는 평균으로 예측'), dataset)
```

```python
plt.figure()
plt.scatter(dataset['sqfeet'], dataset['rent'])
plt.hlines(1418, 750, 950, color = 'black', linestyle = '--', linewidth = 1)
for idx in range(len(dataset)):
    plt.vlines(dataset['sqfeet'].iloc[idx], dataset['rent'].iloc[idx], 1418, color = 'red')

plt.title('Scatter plot and Residual')
plt.xlabel('SqFeet')
plt.ylabel('Rent');
```

```python
from sklearn.tree import DecisionTreeRegressor

'''
학습률 = 1
weak model = regression stump
f_0 = mean(y)
'''
delta_1 = DecisionTreeRegressor(max_depth = 1)
delta_2 = DecisionTreeRegressor(max_depth = 1)
delta_3 = DecisionTreeRegressor(max_depth = 1)

# first weak learner
delta_1.fit(dataset['sqfeet'].values.reshape(-1, 1), dataset['y - $F_0$'].values.reshape(-1, 1))
F_1 = F_0 + delta_1.predict(dataset['sqfeet'].values.reshape(-1, 1))
residual_1 = dataset['rent'] - F_1

# second weak learner
delta_2.fit(dataset['sqfeet'].values.reshape(-1, 1), residual_1)
F_2 = F_1 + delta_2.predict(dataset['sqfeet'].values.reshape(-1, 1))
residual_2 = dataset['rent'] - F_2

# thrid weak learner
delta_3.fit(dataset['sqfeet'].values.reshape(-1, 1), residual_2)
F_3 = F_2 + delta_3.predict(dataset['sqfeet'].values.reshape(-1, 1))

dataset['$\Delta_1$'] = delta_1.predict(dataset['sqfeet'].values.reshape(-1, 1))
dataset['$F_1$'] = F_1
dataset['y - $F_1$'] = residual_1
dataset['$\Delta_2$'] = delta_2.predict(dataset['sqfeet'].values.reshape(-1, 1))
dataset['$F_2$'] = F_2
dataset['y - $F_2$'] = residual_2
dataset['$\Delta_3$'] = delta_3.predict(dataset['sqfeet'].values.reshape(-1, 1))                                   
dataset['$F_3$'] = F_3

display(Markdown('### 3번째 weak leaner까지 학습'), dataset)
```

```python
fig, ax = plt.subplots(1, 3, constrained_layout = True, sharey = True, figsize = (15, 5))
ax[0].scatter(dataset['sqfeet'], dataset['y - $F_0$'], label = 'y - $F_0$')
ax[0].plot(np.arange(750, 951), delta_1.predict(np.arange(750, 951).reshape(-1, 1)), '--k', label = '$\Delta_1(x)$')
ax[0].hlines(0, 750, 950, linestyle = '--', color = 'blue', alpha = .5)
ax[0].set_xlabel('SqFeet', fontsize = 20)
ax[0].set_ylabel('$y - \hat{y}$', fontsize = 20)
ax[0].legend(loc = 2, prop={'size': 25})

ax[1].scatter(dataset['sqfeet'], dataset['y - $F_1$'], label = 'y - $F_1$')
ax[1].plot(np.arange(750, 951), delta_2.predict(np.arange(750, 951).reshape(-1, 1)), '--k', label = '$\Delta_2(x)$')
ax[1].hlines(0, 750, 950, linestyle = '--', color = 'blue', alpha = .5)
ax[1].set_xlabel('SqFeet', fontsize = 20)
ax[1].set_ylabel('$y - \hat{y}$', fontsize = 20)
ax[1].legend(loc = 2, prop={'size': 25})

ax[2].scatter(dataset['sqfeet'], dataset['y - $F_2$'], label = 'y - $F_2$')
ax[2].plot(np.arange(750, 951), delta_3.predict(np.arange(750, 951).reshape(-1, 1)), '--k', label = '$\Delta_3(x)$')
ax[2].hlines(0, 750, 950, linestyle = '--', color = 'blue', alpha = .5)
ax[2].set_xlabel('SqFeet', fontsize = 20)
ax[2].set_ylabel('$y - \hat{y}$', fontsize = 20)
ax[2].legend(loc = 2, prop={'size': 25});
```

- 위의 그래프를 보면 weak model이 추가될 수록 residual값이 줄어들고 있는 것을 볼 수 있다.

```python
x = np.arange(750, 951).reshape(-1, 1)
pred_0 = F_0
pred_1 = F_0 + delta_1.predict(x)
pred_2 = F_0 + delta_1.predict(x) + delta_2.predict(x)
pred_3 = F_0 + delta_1.predict(x) + delta_2.predict(x)

fig, ax = plt.subplots(1, 4, constrained_layout = True, figsize = (20, 5))
ax[0].scatter(dataset['sqfeet'], dataset['rent'])
ax[0].plot(x, np.zeros_like(x)+pred_0, '--k', label = '$f_0$')
ax[0].legend(loc = 2, prop = {'size':20})

ax[1].scatter(dataset['sqfeet'], dataset['rent'])
ax[1].plot(x, pred_1, '--k', label = '$f_0 + \eta\Delta_1$')
ax[1].legend(loc = 2, prop = {'size':20})

ax[2].scatter(dataset['sqfeet'], dataset['rent'])
ax[2].plot(x, pred_2, '--k', label = '$f_0 + \eta(\Delta_1 + \Delta_2)$')
ax[2].legend(loc = 2, prop = {'size':20})

ax[3].scatter(dataset['sqfeet'], dataset['rent'])
ax[3].plot(x, pred_3, '--k', label = '$f_0 + \eta(\Delta_1 + \Delta_2 + \Delta_3)$')
ax[3].legend(loc = 2, prop = {'size':20});
```

- 위의 그래프를 보면 overall model의 회귀선이 weak model이 추가됨에 따라 복잡해지고 있음을 볼 수 있다.

> $f_0 + \eta(\Delta_1 + \Delta_2 + \Delta_3)$을 보면 학습률을 마지막에 곱해줘도 될 것 같다.  
수학적으로 틀리지는 않지만 weak model $\Delta_m$이 $y - F_{m-1}(x)$에 대해 학습하고, $F_{m-1}(x) = F_{m-2}(x) + \eta\Delta_{m-1}(x)$인 사실이 숨어있다. Friedman은 이 것을 **incremental shrinkage**라고 표현했다.

### Measuring model performance
모델이 잘하고 있는지 평가하기 위해서는 손실함수 또는 비용함수를 정의해야 한다. N개 샘플에 대한 손실은 각각의 샘플에 대한 손실의 평균이다.
> $L(y, F_M(X)) = \frac{1}{N}\sum_{i=1}^{N}{L(y_i, F_M(x_i))})$
> 
> **MSE(Mean Squared Error)**  
> $L(y, F_M(X)) = \frac{1}{N}\sum_{i=1}^{N}{(y_i - F_M(x_i))^2}$

### Choosing hyper-parameters
- M: number of stages
    - M이 높을수록 model의 정확도가 올라가지만, 과대적합의 가능성도 올라간다.

  
- $\eta$: learning rate
    - overall model의 과대적합을 줄이는 목적
    
Friedman은 0.1과 같이 낮은 learning rate와 높은 M을 추천한다. 하지만 일반적으로 사람들은 hyper-parameter tuning을 위해 grid-search를 하는 경우가 많다. learning rate scheduling 또한 좋은 방법이 될 수 있다.





### GBM algorithm to minimize L2 loss
